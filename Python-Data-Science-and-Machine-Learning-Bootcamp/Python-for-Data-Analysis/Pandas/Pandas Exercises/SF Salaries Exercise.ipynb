** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
job_counts = sal[sal['Year'] == 2013]['JobTitle'].value_counts()
len(job_counts[job_counts == 1])
```

```python
# From the solutions - sum will only count True values
sum(sal[sal['Year'] == 2013]['JobTitle'].value_counts() == 1)
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
# First pull all the people who have cheif in their job title
cheifs = sal[sal['JobTitle'].apply(lambda title: 'chief' in title.lower())]
cheifs
```

```python
# Names are repeated more than once if that person was in the database for more than one year
# For instance JOSEPH DRISCOLL and Joseph Driscoll both exist in the EmployeeName category
# as JOSEPH DRISCOLL is for 2011 data and Joseph Driscoll is for 2012 data
# Need to count just unique- the solutions in the QA forums don't account for this
cheifs_lowername = cheifs['EmployeeName'].apply(lambda name: name.lower())
cheifs_lowername
```

```python
cheifs_lowername.value_counts()[cheifs_lowername.value_counts() > 1]
```

```python
# And here's our answer
cheifs_lowername.nunique()
```

```python
# Here is the answer from the solutions
sum(sal['JobTitle'].apply(lambda title: 'chief' in title.lower().split()))
```
